### Diabetes Predictions

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data,columns=["age",
    "sex","bmi",'bp','s1','s2','s3','s4','s5','s6'])
```

```python
df_diabetes
```

```python
## Independent features
X= df_diabetes
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y, train_size=0.3, random_state=10)
```

```python
X_train
```

```python
# correlation
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

### Hyperparameter tuning

```python
param = {
    'criterion' :['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter' :['best','random'],
    'max_depth' :[1,2,3,4,5,10,20,25],
    'max_features' : ['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(regressor,param_grid=param, scoring='neg_mean_squared_error',cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print(r2_score(y_pred,y_test))
print(mean_absolute_error(y_pred,y_test))
print(mean_squared_error(y_pred,y_test))
```

```python
selectmodel = DecisionTreeRegressor(criterion = 'friedman_mse',
 max_depth = 3,
 max_features ='sqrt',
 splitter = 'best')
```

```python
selectmodel.fit(X_train,y_train)
```

```python
## visualize the decision tree
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(selectmodel,filled=True)
```

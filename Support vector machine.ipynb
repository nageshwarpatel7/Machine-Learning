## Support Vector Classification Implementaion

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Lets create a synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples = 1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,random_state=10,train_size=0.25)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
# predictions
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train,y_train)
# predictions
y_pred1 = rbf.predict(X_test)
print(classification_report(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
```

```python
poly = SVC(kernel='poly')
poly.fit(X_train,y_train)
# predictions
y_pred2 = poly.predict(X_test)
print(classification_report(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
# predictions
y_pred3 = sigmoid.predict(X_test)
print(classification_report(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

## Hyperparameter Tuning using SVC

```python
## hyperparameter tuning
from sklearn.model_selection import GridSearchCV

## defining parameter range
param_grid = {'C': [0.1,1,10,100,1000],
              'gamma':[1,0.1,0.01,0.001,0.0001],
              'kernel': ['rbf']
              }
```

```python
grid =GridSearchCV(SVC(),param_grid=param_grid,cv=5,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(classification_report(y_test,y_pred4))
```

### SVM Kernel Indepth intution

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5.0, 5.0, 100)
y = np.sqrt(10**2 - x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
np.vstack([y,x]).T
```

```python
import pandas as pd
df1 =pd.DataFrame(np.vstack([y,x]).T,columns=['X1','X2'])
df1['Y']=0
df2 =pd.DataFrame(np.vstack([y1,x1]).T,columns=['X1','X2'])
df2['Y']=1
df = pd.concat([df1,df2])
df.head(5)
```

```python
df.tail()
```

```python
## Independent and dependent features
X = df.iloc[:,:2]
y = df.Y
```

```python
y
```

```python
## divide data into train test split
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
y_train
```

```python
# We need to find components for the Polynomical Kernel
#X1,X2,X1_square,X2_square,X1*X2
df['X1_Square']= df['X1']**2
df['X2_Square']= df['X2']**2
df['X1*X2'] = (df['X1'] *df['X2'])
df.head()
```

```python
### Independent and Dependent features
X = df[['X1','X2','X1_Square','X2_Square','X1*X2']]
y = df['Y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.25, 
                                                    random_state = 0)
```
